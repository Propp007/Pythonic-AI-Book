# Sigmoid function

```python
import matplotlib.pyplot as plt
import numpy as np
def sigmoid(f):
  return 1/(1 + np.exp(-f))

f = np.linspace (-10,10,50)
print(f)
```

```python
g = np.round(sigmoid(f), 3)
g
```

```python
plt.plot(f,g)
plt.xlabel("f")
plt.ylabel("y")
```

# Tanh function

```python
g = np.tanh(f)
print(g)
plt.plot(f,g)
plt.xlabel("f")
plt.ylabel("g")

```

# ReLu function

```python
def relu(f):
  return np.maximum(f,0)

g = relu(f)
plt.plot(f,g)
plt.xlabel("f")
plt.ylabel("g (relu)")
```
